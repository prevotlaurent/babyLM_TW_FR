```python
import re
import os

from tqdm import tqdm

# For wiki
from datasets import load_dataset

import collections

```

```python
#Input FOLDER
CONV_FR_FOLDER = './corpora/fr/conv/'
OS_FR_FOLDER = './corpora/fr/os/'

#Output Files
WIKI_FR_FILE = './data/raw_data/fr/wiki_fr_10M.txt'
CONV_FR_FILE = './data/raw_data/fr/conv_fr_10M.txt'
OS_FR_FILE = './data/raw_data/fr/OS_fr_10M.txt'


SENT_SYMBOLS="\.|\?|!|\n"
COMMENT_SYMBOLS=['#',"%"]
```

```python
CLEANING_DICT = {
        '.':'',
        '?':'',
        ' ,':',',
        '!':'',
        ':':'',
        '@c':'',
        ';':'',
        '[':'',
        ']':'',
        '(':'',
        ')':'',
        '<': '',
        '>': '',
        '+':'',
        '=': '',
        '+//': '',
        '+/': '',
        '/': '',
        '&': '',    
        ' \n':'\n'
}

def clean_sentence(sent):
    tmp = re.sub(r"\[.+\]",r'',sent.lower())
    tmp = re.sub(r"\{.+\}",r'',tmp)
    tmp = re.sub(r"\+<.+>\+",r'',tmp)
    tmp = re.sub(r"<.+>",r'',tmp)
    tmp = re.sub(r"^ ",r'',tmp)
    for k in CLEANING_DICT.keys():
        tmp = tmp.replace(k,CLEANING_DICT[k])
    return tmp

def extract_dataset(foldername_in,filename_out,
                    size_max=10000000,
                    COMMENT_SYMBOLS=['#',"%"],
                    SENT_SYMBOLS="\.|\?|!|\n"):
    
    foutput = open(filename_out,'w')
    total_tokens = 0

    for filename_in in os.listdir(foldername_in):
        print(filename_in)
        if total_tokens > size_max:
            break;
        finput = open(foldername_in+filename_in,'r')
        lines_in = finput.readlines()
        lines_in = [l for l in lines_in if l[0] not in COMMENT_SYMBOLS]
        for line in tqdm(lines_in):
            sents = re.split(SENT_SYMBOLS,line)
            sents = [clean_sentence(s) for s in sents]
            foutput.writelines([s+'\n' for s in sents if s not in ['\n','']])
            total_tokens += sum([len(text.split()) for text in sents])
            if total_tokens > size_max:
                finput.close()
                break;
        print(total_tokens)
    foutput.close()
    return total_tokens
```

```python
# Load a portion of the French Wikipedia
dataset = load_dataset("wikipedia", "20220301.fr", split="train")

foutput = open(WIKI_FR_FILE,'w')
i = 0
total_tokens = 0

while total_tokens < 10000000 :
    sents = re.split(SENT_SYMBOLS, dataset[i]['text'])
    sents = [s for s in sents if s not in ['\n','']]
    foutput.writelines([clean_sentence(s)+'\n' for s in sents if s[0] not in COMMENT_SYMBOLS])
    total_tokens += sum([len(text.split()) for text in sents])
    i+=1
    if i % 100 == 0:
        print(total_tokens)
        print(i)

foutput.close()
```
